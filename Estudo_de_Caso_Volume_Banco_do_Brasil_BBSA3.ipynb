#Análise de volume
Importando as bibliotecas:

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

Baixando informações de Banco do Brasil da biblioteca:

```python
BB = yf.download('BBAS3.SA')

```

Inspecionando e formatando o dataframe

```python
BB = BB.droplevel(1 , axis = 1)
BB.head()
```

Plotando gráfico de fechamento

```python
BB.columns.name = ''
BB.Close.plot();
```

Plotando o gráfico do volume:

```python
BB.Volume.plot(figsize = (12,6));
```

Determinando a área do gráfico para criação de um gráfico de fechamento com volume:

```python
area_grafico = plt.figure(figsize = (15,10))
area = area_grafico.add_gridspec(nrows = 6, ncols = 1)
graf1 = area_grafico.add_subplot(area[0:5, 0])
graf2 = area_grafico.add_subplot(area[5, 0])
graf1.plot(BB['Close'])
graf2.bar(BB.index, BB.Volume)
graf1.set_title('Gráfico de fechamento | Volume');
```
