```python
import yfinance as yf
import pandas as pd
```

```python
FII = ['BRCO11.SA', 'BRCR11.SA', 'BTAL11.SA', 'BTLG11.SA', 'CPTS11.SA', 'DEVA11.SA', 'KNSC11.SA', 'RBRR11.SA', 'VISC11.SA', 'XPML11.SA']
AST = ['ALOS3.SA', 'BBDC4.SA', 'BRAP4.SA', 'BRSR6.SA', 'ISAE4.SA', 'TRPL4.SA', 'JSLG3.SA', 'MILS3.SA', 'MLAS3.SA', 'POSI3.SA', 'SIMH3.SA',
        'TASA4.SA', 'TUPY3.SA', 'VLID3.SA', 'VULC3.SA', 'WIZC3.SA']
IND = ['USDBRL=X', '^BVSP', '^GSPC', 'EURBRL=X', 'XFIX11.SA']
```

```python
FII2 = yf.download(FII, start = '2022-01-01')
```

```python
FII2.head()
```

```python
FII2.columns.names = ['','']
FII2.head()
```

```python
FII2 = FII2.Close
```

```python
FII2.dropna(inplace = True)
FII2.head()
```

```python
FII2
```

```python
Comprados = 1000/FII2.iloc[0]
Comprados = round(Comprados, 0)
Comprados
```

```python
PL_FII = Comprados*FII2
PL_FII.head()
```

```python
PL_FII["Valor_investido"] = round(PL_FII.sum(axis = 1),2)
PL_FII.head()
```

```python
PL_FII.tail()
```

```python
PL_FII.Valor_investido.plot();
```

```python
IFIX = yf.download('XFIX11.SA', start = '2022-01-01')['Close']
IFIX.columns.name = ''
IFIX.head()
```

```python
PL_FII2 = pd.merge(PL_FII, IFIX, how = 'inner', on = 'Date')
PL_FII2.head()
```

```python
PL_FII2.rename({'XFIX11.SA': 'IFIX'}, axis = 1, inplace = True)
PL_FII2.head()
```

```python
PL_FII3 = (PL_FII2/PL_FII2.iloc[0])*100
PL_FII3.head()
```

```python
PL_FII3.plot(figsize = (15,10));
```

```python
PL_FII3[['Valor_investido', 'IFIX']].plot(figsize = (15,10));
```
