```python
import yfinance as yf
import pandas as pd
```

```python
AST = ['ALOS3.SA', 'BBDC4.SA', 'BRAP4.SA', 'BRSR6.SA', 'ISAE4.SA', 'TRPL4.SA', 'JSLG3.SA', 'MILS3.SA', 'MLAS3.SA', 'POSI3.SA', 'SIMH3.SA',
        'TASA4.SA', 'TUPY3.SA', 'VLID3.SA', 'VULC3.SA', 'WIZC3.SA']
```

```python
AST2 = yf.download(AST, start = '2024-01-01')['Close']
```

```python
AST2.columns.name = ''
AST2.head()
```

```python
AST2.drop(columns = ['ISAE4.SA', 'TRPL4.SA'], inplace = True)
AST2
```

```python
TRPL = yf.download('TRPL4.SA', start = '2024-01-01')['Close']
ISAE = yf.download('ISAE4.SA', start = '2024-01-01')['Close']
```

```python
ELETRICAS = pd.merge(TRPL, ISAE, how = 'outer', on = 'Date')
ELETRICAS
```

```python
ELETRICAS['ELETRICAS'] = ELETRICAS['TRPL4.SA'].combine_first(ELETRICAS['ISAE4.SA'])
ELETRICAS = ELETRICAS.drop(columns = ['TRPL4.SA', 'ISAE4.SA'])
ELETRICAS
```

```python
AST3 = pd.merge(AST2, ELETRICAS, how = 'inner', on = 'Date')
AST3.head()
```

```python
comprados = round((1000/AST3.iloc[0]),0)
PL = comprados * AST3
PL['Valor_investido'] = PL.sum(axis = 1)
PL
```

```python
PL['Valor_investido'].plot(figsize = (10,6));
```

```python
Benchmark = yf.download('^BVSP', start = '2024-01-01')['Close']
Benchmark.rename(columns = {'^BVSP':'IBOV'}, inplace = True)
Benchmark['IBOV'] = round(Benchmark['IBOV'],2)
Benchmark
```

```python
comparacao = pd.merge(PL, Benchmark, how = 'inner', on = 'Date')
comparacao.tail()
```

```python
comparacao_normalizado = (comparacao/comparacao.iloc[0])
comparacao_normalizado
```

```python
comparacao_normalizado[['Valor_investido', 'IBOV']].plot(figsize = (10,6));
```
